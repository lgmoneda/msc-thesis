```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.DataFrame()
df['delta_auc'] = -1 * np.array([0.008, 
                                 .002,
                                 0.011,
                                 -0.023,
                                 0.002,
                                -0.013])
df['domain_auc'] = [0.8275605579916265, 
                    0.6135,
                    0.683869215677744,
                    0.8431863913017777,
                    0.683725148559403,
                    0.7938365727342963] 
df['dataset'] = ['Olist', 'Building permits', 'Chicago crimes', 'Kickstarter', 'Animal shelter',
                'GE News']
df.sort_values(by='domain_auc', inplace=True, ascending=False)

ax = sns.barplot(x='dataset', y='domain_auc', data=df, color='green')
ax.set(xlabel='', ylabel='Domain classifier AUC')
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
ax.xaxis.set_tick_params(labelsize=12)
ax.yaxis.set_tick_params(labelsize=12)
plt.savefig("../images/domain_classifier_performance_bars.eps", format="eps")
```

```python
from matplotlib.collections import PolyCollection as p
fig, ax = plt.subplots(figsize=(12, 6))
sns.regplot(x="domain_auc", y="delta_auc", data=df)
plt.xlabel("Domain classifier AUC", fontsize=15)
plt.ylabel("Delta AUC with TRF", fontsize=15)
ax.xaxis.set_tick_params(labelsize=12)
ax.yaxis.set_tick_params(labelsize=12)

__file__ = "../images/domain_classifier_regression"
plt.savefig(__file__+".jpg", pil_kwargs={'quality': 95}, bbox_inches="tight")

ax.set_rasterized(True)
plt.savefig(__file__ + ".eps", format="eps", bbox_inches="tight")
plt.show()
```

## AUC improvement by Support match 

```python
df = pd.DataFrame()
df['delta_auc'] = -1 * np.array([0.008, 
                                 .002,
                                 0.011,
                                 -0.023,
                                 0.002])
df['support_match'] = [0.7839734229965991, 
                    0.8864631704373713,
                    0.8488993119305348,
                    0.6111019736842105,
                    0.8591957677715395] 
df['dataset'] = ['Olist', 'Building permits', 'Chicago crimes', 'Kickstarter', 'Animal shelter']
df.sort_values(by='support_match', inplace=True, ascending=False)

ax = sns.barplot(x='dataset', y='support_match', data=df, color='green')
ax.set(xlabel='', ylabel='Domain classifier AUC')
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
ax.xaxis.set_tick_params(labelsize=12)
ax.yaxis.set_tick_params(labelsize=12)
plt.savefig("../images/support_match_bars.eps", format="eps")
```

```python
from matplotlib.collections import PolyCollection as p
fig, ax = plt.subplots(figsize=(12, 6))
sns.regplot(x="support_match", y="delta_auc", data=df)
plt.xlabel("Support match", fontsize=15)
plt.ylabel("Delta AUC with TRF", fontsize=15)
ax.xaxis.set_tick_params(labelsize=12)
ax.yaxis.set_tick_params(labelsize=12)

__file__ = "../images/support_match_regression"
plt.savefig(__file__+".jpg", pil_kwargs={'quality': 95}, bbox_inches="tight")

ax.set_rasterized(True)
plt.savefig(__file__ + ".eps", format="eps", bbox_inches="tight")
plt.show()
```
